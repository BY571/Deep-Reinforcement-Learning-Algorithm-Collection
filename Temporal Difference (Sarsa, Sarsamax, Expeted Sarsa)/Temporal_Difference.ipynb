### Part 1: TD Control: Sarsa

In this section, you will write your own implementation of the Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def probs(actions, eps, nA):
    prob = np.ones(nA)* eps/nA
    action = np.argmax(actions)
    prob[action] = 1 - eps + (eps/nA)
    return prob
    
def update_Q(Q, state, action, reward, new_state,done, alpha, gamma, nA, eps):
    new_action = np.random.choice(np.arange(nA), p = probs(Q[new_state], eps, nA)) \
                                    if new_state in Q else env.action_space.sample()
    if not done:
        Q_next = Q[new_state][new_action]
    else:
        Q_next = 0
    Q[state][action] = Q[state][action] + alpha * (reward + gamma * Q_next - Q[state][action])
    return Q, new_action
````

````python
def run_optimization(env, Q, eps, nA, alpha, gamma):
    state = env.reset()
    episodes = []
    rewards = []
    done = False
    action = np.random.choice(np.arange(nA), p = probs(Q[state], eps, nA)) \
                                    if state in Q else env.action_space.sample()
    while not done:
        new_state, reward, done, _ = env.step(action)
        if done:
            Q, _ = update_Q(Q, state, action, reward, new_state,done, alpha, gamma, nA, eps)
            break
        Q, new_action = update_Q(Q, state, action, reward, new_state,done, alpha, gamma, nA, eps)
        rewards.append(reward)
        state = new_state
        action = new_action

    return Q, rewards
````

````python
def sarsa(env, num_episodes, eps_init = 1.0, eps_decay = 0.95, eps_min = 0.0025, alpha = 0.009, gamma = 1.0):
    # initialize action-value function (empty dictionary of arrays)
    nA = env.nA
    Q = defaultdict(lambda: np.zeros(nA))
    # initialize performance monitor
    # loop over episodes
    rewards = []
    eps = eps_init
    for i_episode in range(1, num_episodes+1):  
        eps = max(eps*eps_decay, eps_min)
        Q, R = run_optimization(env, Q, eps, nA, alpha, gamma)
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush() 

        rewards.append(np.mean(R))
    plt.plot(rewards)
    plt.xlabel("Episodes")
    plt.ylabel("Rewards")
    
    return Q
````

### Part 2: TD Control: Q-learning

In this section, you will write your own implementation of the Q-learning control algorithm.

Your algorithm has four arguments:

- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def probs(actions, eps, nA):
    prob = np.ones(nA)* eps/nA
    action = np.argmax(actions)
    prob[action] = 1 - eps + (eps/nA)
    return prob

def update_Q(Q, state, action, reward, new_state, alpha, gamma, nA, eps):
    new_action = np.argmax(Q[new_state])
    Q[state][action] = Q[state][action] + alpha * (reward + gamma * Q[new_state][new_action] - Q[state][action])
    return Q
````

````python
def run_optimization(env, Q, eps, nA, alpha, gamma):
    state = env.reset()
    episodes = []
    done = False
    while not done:
        action = np.random.choice(np.arange(nA), p = probs(Q[state], eps, nA)) \
                                    if state in Q else env.action_space.sample() #np.argmax(Q[state])
        new_state, reward, done, _ = env.step(action)
        Q = update_Q(Q, state, action, reward, new_state, alpha, gamma, nA, eps)
        state = new_state
        if done:
            break
    return Q
````

````python
def q_learning(env, num_episodes, eps_init = 1.0, eps_decay = 0.99, eps_min = 0.05, alpha = 0.01, gamma=1.0):
    # initialize empty dictionary of arrays
    nA = env.nA
    Q = defaultdict(lambda: np.zeros(env.nA))
    eps = eps_init
    #loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        eps = max(eps*eps_decay, eps_min)
        Q = run_optimization(env, Q, eps, nA, alpha, gamma)
        
    return Q
````

### Part 3: TD Control: Expected Sarsa

In this section, you will write your own implementation of the Expected Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def probs(actions, eps, nA):
    prob = np.ones(nA)* eps/nA
    action = np.argmax(actions)
    prob[action] = 1 - eps + (eps / nA)
    return prob

def update_Q(Q, state, action, reward, new_state, done, alpha, gamma, nA, eps):
    probabilities = probs(Q[new_state], eps, nA)
    actions = np.arange(nA)
    best_action = actions[np.argmax(probabilities)]
    Q_target = (1-done)*(sum([probabilities[action] * Q[new_state][action] for action in actions if action != best_action]) \
                + probabilities[best_action]*Q[new_state][best_action])
    
    Q[state][action] = Q[state][action] + alpha * (reward + gamma * Q_target  - Q[state][action])
    return Q
````

````python
def run_optimization(env, Q, eps, nA, alpha, gamma):
    state = env.reset()
    episodes = []
    rewards = []
    done = False
    action = np.random.choice(np.arange(nA), p = probs(Q[state], eps, nA)) \
                                    if state in Q else env.action_space.sample() #np.argmax(Q[state])
    while not done:
        
        new_state, reward, done, _ = env.step(action)
        rewards.append(reward)
        new_action = np.random.choice(np.arange(nA), p = probs(Q[new_state], eps, nA)) \
                                    if new_state in Q else env.action_space.sample()
    
        Q = update_Q(Q, state, action, reward, new_state, done, alpha, gamma, nA, eps)
        state = new_state
        action = new_action
        if done:
            break
    return Q, rewards
````

````python
def expected_sarsa(env, num_episodes, eps_init = 1.0, eps_decay = 0.95, eps_min = 0.0025, alpha = 0.01, gamma=1.0):
    # initialize empty dictionary of arrays
    nA = env.nA
    Q = defaultdict(lambda: np.zeros(nA))
    # loop over episodes
    eps = eps_init
    rewards = []
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        eps = max(eps*eps_decay, eps_min)
        Q,R = run_optimization(env, Q, eps, nA, alpha, gamma)
        rewards.append(np.mean(R))
    plt.plot(rewards)
    plt.xlabel("Episodes")
    plt.ylabel("Rewards")
    return Q
````
